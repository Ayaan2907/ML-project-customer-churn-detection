# Column Separation

```python
feature = df.drop('Exited', axis=1)
feature
```

```python
target = df['Exited']
target
```

```python
feature = feature.drop(['CustomerId', 'RowNumber', 'Surname' ], axis=1)
feature
```

```python
# from google.colab import sheets
# sheet = sheets.InteractiveSheet(df=feature)
```

```python
feature = feature.dropna() #in case missing data, drop it ( u have to deal it case by case in other projects )
feature
```

```python
# encoding categorical features, models only understand numerics, boolen, so to
# convert into onehardcoding we use this
features = pd.get_dummies(feature, columns=['Geography', 'Gender'])
features
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train[0]
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression

from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb


import pickle # used to save ML models
```

```python
lr_model = LogisticRegression(random_state=42)
lr_model.fit(X_train, y_train)
```

```python
lr_predications = lr_model.predict(X_test)
lr_predications
```

```python
accuracy = accuracy_score(y_test, lr_predications)
accuracy
```

# Note

We are training our model (first training iteration)



```python
def evaluate_and_save_model(model, X_train, X_test, y_train, y_test, filename):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print("-------------------------")
    print(f"{model.__class__.__name__}: Accuracy = {accuracy:.4f}")
    print(f"Classification Report : \n {classification_report(y_test, y_pred)}")
    print(confusion_matrix(y_test, y_pred))

    with open(filename, 'wb') as file:
        pickle.dump(model, file)
    print(f"Model saved as {filename}")
```

```python
xgb_model = xgb.XGBClassifier(random_state=42)
evaluate_and_save_model(xgb_model, X_train, X_test, y_train, y_test, 'models/xgb_model.pkl')

dt_model = DecisionTreeClassifier(random_state=42)
evaluate_and_save_model(dt_model, X_train, X_test, y_train, y_test, 'models/dt_model.pkl')

rf_model = RandomForestClassifier(random_state=42)
evaluate_and_save_model(rf_model, X_train, X_test, y_train, y_test, 'models/rf_model.pkl')

nb_model = GaussianNB()
evaluate_and_save_model(nb_model, X_train, X_test, y_train, y_test, 'models/nb_model.pkl')

knn_model = KNeighborsClassifier()
evaluate_and_save_model(knn_model, X_train, X_test, y_train, y_test, 'models/knn_model.pkl')

svm_model = SVC(random_state=42, probability=True)
evaluate_and_save_model(svm_model, X_train, X_test, y_train, y_test, 'models/svm_model.pkl')
```

# Improvements

  1. Feature Engineering
    Creating new features on the dataset based on existing features to help better train the model

  2. SMOTE (Synthetic Minority Over-sampling TEchnique)
    a technique used in machine learning to address class imbalance in datasets by artificially increasing the number of minority class samples

  3. Ensembeling
    This is like asking multiple experts their opinion and then taking a majority vote, process of combining output of multiple models to get a new model trained.

  > Hyper parameter tuning, but not for this now

```python
# feature engineering
# lets add new features (lets add CLV : customer long term value, a multiple of balance and salary)
feature['CLV'] = df['Balance'] * df['EstimatedSalary']/100000
feature
# added new columns CLV
```

```python
# lets group our features ( age )
feature['AgeGroup'] = pd.cut(df['Age'], bins=[0,30,45, 60, 100], labels=['young', 'middle', 'senior', 'old'])
feature
```

```python
# adding new feature
feature['TenureAgeRatio'] = df['Tenure'] / df['Age']
feature
```

```python
# Now we again need to  convert categorical variables to numeric or boolen
feature = pd.get_dummies(feature, drop_first=True)
feature
```

```python
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, random_state=42)
```

```python
# scale the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
xgboost_model = xgb.XGBClassifier(random_state=42)
evaluate_and_save_model(xgboost_model, X_train, X_test, y_train, y_test, 'models/xgboost_featureEng_model.pkl')
```

# ENsambeling

voting classifier from sklearn which combines voting from multiple models

```python
from sklearn.ensemble import VotingClassifier
```

```python
voting_clf = VotingClassifier(estimators=[
    ('xgb', xgb.XGBClassifier(random_state=42)),
    ('dt', DecisionTreeClassifier(random_state=42)),
    ('rf', RandomForestClassifier(random_state=42)),
    ('nb', GaussianNB()),
    # `('knn', KNeighborsClassifier()),
    ('svm', SVC(random_state=42, probability=True))
], voting='soft')
```

```python
evaluate_and_save_model(voting_clf, X_ressampled, X_test, Y_resampled, y_test, 'models/voting_clf.pkl')
# this is done by setting the voting param HARD
```
